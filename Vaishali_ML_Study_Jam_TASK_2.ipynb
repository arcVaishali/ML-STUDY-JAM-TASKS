#### IMPORTING PANDAS AND LOADING CSV FILE

```python
import pandas as pd

df = pd.read_csv("/content/ML_study_jam.csv")
print(df)
```

## **a) Remove rows/ attributes**

```python
#REMOVING ROWS---
print("REMOVING ROW WITH INDEX-1")
df = df.drop(1)
print(df)
```

```python
#REMOVING COLUMNS----
print("REMOVING COLUMN WITH LABEL Store_Area") 
df = df.drop('Store_Area', axis = 1) #passing column name and axis parameter=1(for columns)
print(df)
```

```python
#WRITING THE MODIFIED DATAFRAME BACK TO CSV FILE
df.to_csv("/content/ML_study_jam.csv", index=False)  

print(df)
```

#### REMOVING ROWS/ATTRIBUTES WITH MISSING VALUES

```python
#REMOVING ROWS WITH MISSING VALUES----
print("Removing rows with empty values---") 
df2 = df.dropna()  #deleting any row with atleast one missing value and storing the resultant df in df2 

print("MODIFIED DATAFRAME---\n\n")
print(df2)    

print("ORIGINAL DATAFRAME---\n\n")
print(df)
```

```python
# REMOVING COLUMNS WITH MISSING VALUES---
print("Removing columns with missing values  ")
df1 = df.dropna(axis = 1) #deleting any column with atleast one missing value and storing the resultant df in df1

print('\n\nPRINTING MODIFIED DATAFRAME')
print(df1) 

print("\n\nPRINTING ORIGINAL DATAFRAME")
print(df)
```

## **(b.) Replacing missing values in the dataframe with mean/mode** 

```python
print("REPLACING MISSING VALUES WITH MEAN")
meanVal = df.mean()
df3 = df.fillna( meanVal )

df3
```

```python
print("REPLACING MISSING VALUES WITH MODE")
modeval = df.mode().iloc[0]
df4 = df.fillna(modeval)

df4
```

## **(c.) Write a program to remove /replace missing values beyond a threshold value.**

```python
df6 = df.copy()
threshold = 0.5
for i in range(len(df6)):
    if sum( df6.iloc[i].isnull() ) > len( df6.columns )*threshold :
        df6.iloc[i] = df6.iloc[i].fillna(df6.iloc[i].mean()) 

df6
```
